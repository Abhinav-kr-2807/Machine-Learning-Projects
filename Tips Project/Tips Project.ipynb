About Dataset:

After receiving satisfactory service, it is customary to add a small gratuity to the bill and give it to the restaurant personnel. Although tips are optional and generally not indicated, it is customary to leave a percentage of the overall cost (for example, 10%) even though they are optional and not required.

The tips dataset, which is being examined in this Notebook, represents tips left at restaurants. The information in question consists of a record of tips provided together with the total bills of restaurant patrons that was reportedly gathered by a server over the course of several weeks of employment there.


The information is set up in a form of an array, where the observations (also known as instances) are arranged in rows and the dataset properties (also known as features) in columns. The dataset consists of many data types that describe the tips that were given in relation to various factors, such as the day of the week, the tipper's gender, etc. 244 data observations are part of the data set.
